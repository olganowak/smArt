```python
df = df.rename(columns = {0: "genre", 1: "artist", 2: "title"})
```

```python
def replace_underscore(x):
    return x.replace("_", " ")
```

```python
df["genre"] = df.genre.apply(replace_underscore)
```

```python
def capitalizer(x):
    lista = x.split()
    for i, word in enumerate(lista):
        lista[i] = word.capitalize()
    return " ".join(lista)
```

```python
capitalizer("joan mitchell")
```

```python
df["artist"] = df.artist.apply(capitalizer)
```
