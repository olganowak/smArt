```python
df = df.rename(columns = {0: "genre", 1: "artist", 2: "title"})
```

```python
def replace_underscore(x):
    return x.replace("_", " ")
```

```python
df["genre"] = df.genre.apply(replace_underscore)
```

```python
def capitalizer(x):
    if x:
        lista = x.split()
        for i, word in enumerate(lista):
            lista[i] = word.capitalize()
        return " ".join(lista)
```

```python
capitalizer("joan mitchell")
```

```python
df["artist"] = df.artist.apply(capitalizer)
```

```python
df.tail()
```

```python
def year(x):
    if x:
        lista = x.split()
        if len(lista[-1]) == 4 and lista[-1].isdigit():
            return lista[-1]
        return None
```

```python
df["year"] = df.title.apply(year)
df[df["year"].notnull()]
```

```python
def remove_year(x):
    if x:
        lista = x.split()
        if len(lista[-1]) == 4 and lista[-1].isdigit():
            return " ".join(lista[:-1])
        return None
```

```python
df.title = df.title.apply(remove_year)
```
